```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("Student_Grade.csv")
#importing the data
X=data.iloc[:,0]
Y=data.iloc[:,1]
#building the model y=mx+c
m=0
c=0
#learning rate
L=0.000001
epochs=1000
n=float(len(X))
for i in range(len(X)):
    Y_pred=m*X+c
    D_m=(-1/n)*sum(X*(Y-Y_pred))
    D_c=(-1/n)*sum(Y-Y_pred)
    m=m-L*D_m
    c=c-L*D_c
print(m,c)

#making prediction
Y_pred=m*X+c
plt.xlabel=('hours of studying')
plt.ylabel=('grade obtained')
plt.scatter(X,Y,s=10)
plt.scatter(X,Y_pred,s=15)
plt.plot(X,Y_pred,'k')
plt.legend(['actual','predicted','hypothesis function'])
plt.show()
```
